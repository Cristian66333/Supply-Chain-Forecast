```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Asume que tu DataFrame 'df' está cargado, con columnas de medicamentos y 97 filas (valores temporales)
import pandas as pd

df = pd.read_excel(path)
df.head(5)
def crear_dataset(dataset, look_back=3):
    X, Y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:i+look_back, 0])
        Y.append(dataset[i+look_back, 0])
    return np.array(X), np.array(Y)

look_back = 24
resultados = {}
targets = ['CLORURO DE SODIO', 'LACTATO DE SODIO', 'METAMIZOL', 'PARACETAMOL', 'PROPOFOL']
for med in targets:
    print(f"Procesando {med}...")

    serie = df[med].values.astype(float)

    # 1. Ajustar ARIMA
    config_list ={
    'CLORURO DE SODIO': (3, 1, 3),
    'LACTATO DE SODIO': (3, 0, 3),
    'METAMIZOL': (3, 1, 3),
    'PARACETAMOL': (2, 1, 2),
    'PROPOFOL': (3, 0, 0)}

    model_arima = ARIMA(serie, order=config_list[med])
    arima_fit = model_arima.fit()
    pred_arima = arima_fit.predict(start=0, end=len(serie)-1)

    # 2. Calcular residuos y escalar
    residuos = serie - pred_arima
    scaler = MinMaxScaler(feature_range=(0,1))
    res_scaled = scaler.fit_transform(residuos.reshape(-1,1))

    # 3. Preparar datos LSTM
    X, y = crear_dataset(res_scaled, look_back)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # 4. Crear y entrenar LSTM
    model_lstm = Sequential()
    model_lstm.add(LSTM(100, activation='relu', input_shape=(look_back,1)))
    model_lstm.add(Dense(1))
    model_lstm.compile(optimizer='adam', loss='mse')
    model_lstm.fit(X, y, epochs=50, batch_size=1, verbose=0)

    # 5. Predecir con LSTM y desescalar
    pred_lstm_scaled = model_lstm.predict(X)
    pred_lstm = scaler.inverse_transform(pred_lstm_scaled)

    # 6. Combinar predicciones
    pred_arima_trimmed = pred_arima[look_back:]
    pred_final = pred_arima_trimmed + pred_lstm.flatten()

    # Valores reales ajustados (para comparación)
    reales = serie[look_back:]

    # 7. Calcular métricas
    mse = mean_squared_error(reales, pred_final)
    mae = mean_absolute_error(reales, pred_final)
    rmse = np.sqrt(mse)
    r2 = r2_score(reales, pred_final)

    resultados[med] = {
        "mse": mse,
        "mae": mae,
        "rmse": rmse,
        "r2": r2,
        "pred_final": pred_final,
        "reales": reales,
        "pred_arima": pred_arima_trimmed,
        "pred_lstm": pred_lstm.flatten()
    }

    # 8. Graficar resultados
    plt.figure(figsize=(10,5))
    plt.plot(reales, label='Real')
    plt.plot(pred_arima_trimmed, label='Pred ARIMA')
    plt.plot(pred_lstm.flatten(), label='Pred LSTM (residuos)')
    plt.plot(pred_final, label='Pred Híbrida ARIMA+LSTM', linewidth=2)
    plt.title(f'Forecasting medicamento: {med}')
    plt.legend()
    plt.show()

# Mostrar resumen de métricas para todos los medicamentos
print("Métricas para cada medicamento:")
for med, met in resultados.items():
    print(f"{med}: MSE={met['mse']:.3f}, MAE={met['mae']:.3f}, RMSE={met['rmse']:.3f}, R2={met['r2']:.3f}")
```
